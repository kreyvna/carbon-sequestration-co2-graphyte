# Carbon Sequestration Analysis from Mauna Loa CO₂ Data

This project explores atmospheric CO₂ trends using NOAA’s Mauna Loa dataset. 
The goal is to simulate how environmental sensor data can be used to validate carbon sequestration effectiveness.

We will:
- Clean and analyze the CO₂ dataset
- Visualize long-term and seasonal trends
- Apply statistical methods to assess dynamics
- Consider implications for carbon removal validation


```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set style
sns.set(style="whitegrid")
```

```python
# Define file path
file_path = '../data/raw/co2_mm_mlo.csv'

# Count how many comment lines to skip
with open(file_path, 'r') as f:
    skip = 0
    for line in f:
        if line.startswith('#'):
            skip += 1
        else:
            break

# Load the dataset, skipping comment lines
df = pd.read_csv(file_path, skiprows=skip, header=0)

# Assign column names from NOAA excel file
df.columns = ['year', 'month', 'decimal_date', 'average', 'deseasonalized', 'ndays', 'sdev', 'unc']

# Clean up column names
df.columns = df.columns.str.strip().str.lower()

# Convert relevant columns to numeric and handle missing placeholders
df['average'] = pd.to_numeric(df['average'], errors='coerce').replace(-99.99, pd.NA)
df['deseasonalized'] = pd.to_numeric(df['deseasonalized'], errors='coerce').replace(-99.99, pd.NA)

# Drop rows with missing values needed for plotting or modeling
df = df.dropna(subset=['average', 'deseasonalized'])

# Preview structure
df.head()
```

```python
#Use library for font error
import matplotlib as mpl

# Set a default font to avoid Arial warnings
mpl.rcParams['font.family'] = 'DejaVu Sans'  #matplotlib-safe font

# Create CO₂ trend plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='decimal_date', y='average', label='Monthly Average')
sns.lineplot(data=df, x='decimal_date', y='deseasonalized', label='Smoothed Trend')

plt.title('Mauna Loa CO₂ Concentration Over Time')
plt.xlabel('Year')
plt.ylabel('CO₂ (ppm)')
plt.legend()
plt.tight_layout()

# Save plot to file
plt.savefig('../visuals/co2_trend.png')
plt.show()
```

```python
#Import libraries for regression modeling
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data
X = df['decimal_date'].values.reshape(-1, 1)
y = df['average'].values

# Fit model
model = LinearRegression()
model.fit(X, y)

# Add predictions to DataFrame
df['regression_line'] = model.predict(X)

# Display the slope (ppm/year)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression Line: CO₂ = {slope:.2f} × Year + {intercept:.2f}")
```

```python
#Adding regression to plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='decimal_date', y='average', data=df, label='Observed', alpha=0.5)
sns.lineplot(x='decimal_date', y='regression_line', data=df, color='red', label='Linear Fit')

plt.title('Linear Regression of Atmospheric CO₂ Levels')
plt.xlabel('Decimal Year')
plt.ylabel('CO₂ (ppm)')
plt.legend()
plt.tight_layout()
plt.savefig('../visuals/co2_regression.png')
plt.show()
```

```python
# Calculate residuals and z-scores
df['residual'] = df['average'] - df['regression_line']
df['z_score'] = (df['residual'] - df['residual'].mean()) / df['residual'].std()

# Flag anomalies with |z| > 2.5
df['is_anomaly'] = df['z_score'].abs() > 2.5

# Show top anomalies
df[df['is_anomaly']].head()
```

```python
#Highlight anomalies on plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='decimal_date', y='average', data=df, label='Average CO₂')
sns.scatterplot(x='decimal_date', y='average', data=df[df['is_anomaly']], color='red', label='Anomalies')

plt.title('Detected CO₂ Anomalies')
plt.xlabel('Year')
plt.ylabel('CO₂ (ppm)')
plt.legend()
plt.tight_layout()
plt.savefig('../visuals/co2_anomalies.png')
plt.show()
```

```python
# Flag interpolated or incomplete data (NOAA marks these with negative uncertainty)
df['is_interpolated'] = df['unc'] < 0

# Flag early data from Scripps Institution (before NOAA took over)
df['is_scripps'] = df['decimal_date'] < 1974.33  # April 1974

# Flag temporary Maunakea station data (Dec 2022 to July 2023)
df['is_maunakea'] = df['decimal_date'].between(2022.92, 2023.5)
```

```python
plt.figure(figsize=(12, 6))

# Base CO₂ line
sns.lineplot(data=df, x='decimal_date', y='average', label='Observed CO₂')

# Highlight interpolated
sns.scatterplot(data=df[df['is_interpolated']], x='decimal_date', y='average', color='red', label='Interpolated')

# Highlight Maunakea
sns.scatterplot(data=df[df['is_maunakea']], x='decimal_date', y='average', color='orange', label='Maunakea Station')

plt.title('CO₂ Values with Interpolated and Maunakea Periods Highlighted')
plt.xlabel('Year')
plt.ylabel('CO₂ (ppm)')
plt.legend()
plt.tight_layout()
plt.savefig('../visuals/co2_interpolated_and_maunakea.png')
plt.show()
```

```python
#Sumarizing points of concern
print("Total interpolated months:", df['is_interpolated'].sum())
print("Total anomalies detected:", df['is_anomaly'].sum())
print("Total Maunakea Observatories months:", df['is_maunakea'].sum())
```

## Summary of Findings

- Atmospheric CO₂ concentrations have increased at an average rate of **~1.65 ppm per year** since 1958.
- The seasonal fluctuations in CO₂ reflect Earth’s natural carbon cycle — and the deseasonalized trend confirms persistent accumulation.
- Several data points were flagged as anomalies using statistical outlier detection (z-scores).
- Interpolated months were identified using NOAA’s uncertainty flag, and early records from the Scripps Institution were noted.
- This analysis mirrors how climate tech companies like **Graphyte** might monitor, model, and validate the **permanence of carbon removal** using real-world sensor data.

## Data Source Considerations

- Data from 1958 to 1974 was collected by the Scripps Institution of Oceanography and integrated into NOAA's dataset.
- Some monthly CO₂ values are interpolated, indicated by negative uncertainty values.
- A small portion of data (Dec 2022–July 2023) came from a temporary Maunakea station due to volcanic activity.
- These factors were not excluded from this analysis, but they may affect precision in time series modeling and anomaly detection.
